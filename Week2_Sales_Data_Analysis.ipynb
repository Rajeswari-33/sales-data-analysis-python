```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

data = {
    "Order_ID": range(1, 51),
    "Order_Date": pd.date_range(start="2024-01-01", periods=50, freq="D"),
    "Product": np.random.choice(
        ["Laptop", "Mobile", "Tablet", "Headphones", "Smartwatch"], 50
    ),
    "Category": np.random.choice(
        ["Electronics", "Accessories"], 50
    ),
    "Quantity": np.random.randint(1, 6, 50),
    "Unit_Price": np.random.randint(500, 50000, 50)
}

df = pd.DataFrame(data)
df.head()
```

```python
df["Total_Price"] = df["Quantity"] * df["Unit_Price"]
df.head()
```

```python
total_revenue = df["Total_Price"].sum()
total_orders = df["Order_ID"].count()

print("Total Revenue:", total_revenue)
print("Total Orders:", total_orders)
```

```python
product_sales = df.groupby("Product")["Total_Price"].sum().sort_values(ascending=False)
product_sales
```

```python
category_sales = df.groupby("Category")["Total_Price"].sum()
category_sales
```

```python
df["Month"] = df["Order_Date"].dt.month

monthly_sales = df.groupby("Month")["Total_Price"].sum()
monthly_sales
```

```python
df["Order_Date"] = df["Order_Date"].dt.strftime("%Y-%m-%d")
df.head()
```

```python
df.to_csv("week2_sales_analysis_output.csv", index=False)
```
